```python
import os
import pandas as pd 
import numpy as np
from sklearn.preprocessing import MinMaxScaler
from sklearn.decomposition import PCA
from sklearn.discriminant_analysis import LinearDiscriminantAnalysis as LDA
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score, classification_report
from sklearn.metrics import r2_score, mean_squared_error, accuracy_score, classification_report, accuracy_score, precision_score, recall_score, f1_score
from sklearn.metrics import confusion_matrix
import matplotlib.pyplot as plt
from sklearn import tree
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier
from sklearn.naive_bayes import GaussianNB

file_path = "veri-seti.txt"
data = pd.read_csv(file_path, delimiter = "\t")

data
```

```python
df = pd.DataFrame(data)

# Bir min-max scaler'Ä± oluÅŸturuyoruz.
scaler = MinMaxScaler()

#Normalize edeceÄŸimiz sÃ¼tunlarÄ± seÃ§iyoruz.
data_to_normalize = data.copy()

#Normalizasyon iÅŸlemi.
normalized_data = scaler.fit_transform(data_to_normalize)

#NormalleÅŸtirilmiÅŸ veriyi bir dataframe'e dÃ¶nÃ¼ÅŸtÃ¼rÃ¼yoruz.
normalized_df = pd.DataFrame(normalized_data, columns = data_to_normalize.columns)

normalized_df
```

```python
#Target deÄŸiÅŸkenimizi ayÄ±rÄ±yoruz.
X = normalized_df.drop('X', axis=1)
y = normalized_df['X']

#PCA iÅŸleminin uygulanmasÄ±
pca = PCA(n_components=2)
X_pca = pca.fit_transform(X)

#LDA iÅŸleminin uygulanmasÄ±
lda = LDA(n_components=1)
X_lda = lda.fit_transform(X,y)

pca_df = pd.DataFrame(data=X_pca, columns=['PC1', 'PC2'])
lda_df = pd.DataFrame(data=X_lda, columns=['LD1'])
```

```python
print("PCA DeÄŸerleri: ")
print(pca.components_)
print("LDA DeÄŸerleri")
print(lda.scalings_)
```

```python
X = normalized_df.drop('X',axis=1)
y = normalized_df['X']

#Veri setini %70 eÄŸitim ve %30 test verisi olarak ayÄ±rma.
X_train, X_test, y_train, y_test = train_test_split(X,y, test_size = 0.3, random_state = 42)
```

```python
#Ã‡oklu DoÄŸrusal Regresyon modelini oluÅŸturma ve eÄŸitme
linear_reg_model = LinearRegression()
linear_reg_model.fit(X_train, y_train)

# KatsayÄ±larÄ± ve kesme terimini alarak raporlama
linear_reg_coefficients = linear_reg_model.coef_
linear_reg_intercept = linear_reg_model.intercept_

print("Ã‡oklu DoÄŸrusal Regresyon KatsayÄ±larÄ±:")
for i, coef in enumerate(linear_reg_coefficients):
    print(f"KatsayÄ± {i+1}: {coef}")
print(f"Kesme terimi: {linear_reg_intercept}")
```

```python
# Multinominal Lojistik Regresyon modelini oluÅŸturma ve eÄŸitme
logistic_reg_model = LogisticRegression(multi_class='multinomial', solver='lbfgs')
logistic_reg_model.fit(X_train, y_train)

# KatsayÄ±larÄ± alarak raporlama
logistic_reg_coefficients = logistic_reg_model.coef_

print("Multinominal Lojistik Regresyon KatsayÄ±larÄ±:")
for i, coef in enumerate(logistic_reg_coefficients[0]):
    print(f"KatsayÄ± {i+1}: {coef}")
```

```python
# Ã‡oklu DoÄŸrusal Regresyon modeli ile test setinde tahmini
linear_reg_predictions = linear_reg_model.predict(X_test)

# R^2 (R Kare) deÄŸeri
linear_reg_r2 = r2_score(y_test, linear_reg_predictions)
print("Ã‡oklu DoÄŸrusal Regresyon R^2 (R Kare) DeÄŸeri:", linear_reg_r2)

# Ortalama Kare HatasÄ± (MSE)
linear_reg_mse = mean_squared_error(y_test, linear_reg_predictions)
print("Ã‡oklu DoÄŸrusal Regresyon MSE (Ortalama Kare Hata):", linear_reg_mse)
```

```python
# Multinominal Lojistik Regresyon modeli ile test setinde tahmini
logistic_reg_predictions = logistic_reg_model.predict(X_test)

# DoÄŸruluk (Accuracy) hesaplama
logistic_reg_accuracy = accuracy_score(y_test, logistic_reg_predictions)
print("Multinominal Lojistik Regresyon DoÄŸruluk (Accuracy):", logistic_reg_accuracy)

# SÄ±nÄ±flandÄ±rma raporu (classification report) oluÅŸturma
logistic_reg_classification_report = classification_report(y_test, logistic_reg_predictions)
print("Multinominal Lojistik Regresyon SÄ±nÄ±flandÄ±rma Raporu:\n", logistic_reg_classification_report)
```

```python
print("Ã‡oklu DoÄŸrusal Regresyon Modeli PerformansÄ±:")
print("R^2 DeÄŸeri:", linear_reg_r2)
print("Ortalama Kare Hata (MSE):", linear_reg_mse)
print("\nMultinominal Lojistik Regresyon Modeli PerformansÄ±:")
print("DoÄŸruluk (Accuracy):", logistic_reg_accuracy)
print("\nSÄ±nÄ±flandÄ±rma Raporu:")
print(logistic_reg_classification_report)
```

```python
# Multinominal Lojistik Regresyon modeli iÃ§in confusion matrix'i hesaplama
conf_matrix = confusion_matrix(y_test, logistic_reg_predictions)

# Sensitivity ve specificity deÄŸerlerini hesaplama
true_negatives = conf_matrix[0, 0]
false_negatives = conf_matrix[1, 0]
true_positives = conf_matrix[1, 1]
false_positives = conf_matrix[0, 1]

sensitivity = true_positives / (true_positives + false_negatives)
specificity = true_negatives / (true_negatives + false_positives)

# SonuÃ§larÄ± yazdÄ±rma
print("Sensitivity:", sensitivity)
print("Specificity:", specificity)
```

```python
X_pca_train, X_pca_test, y_train, y_test = train_test_split(X_pca, y, test_size=0.3, random_state=42)

# Ã‡oklu DoÄŸrusal Regresyon modelini eÄŸitme ve tahmin
linear_reg_pca = LinearRegression()
linear_reg_pca.fit(X_pca_train, y_train)
linear_reg_pca_preds = linear_reg_pca.predict(X_pca_test)

# Multinominal Lojistik Regresyon modelini eÄŸitme ve tahmin
logistic_reg_pca = LogisticRegression()
logistic_reg_pca.fit(X_pca_train, y_train)
logistic_reg_pca_preds = logistic_reg_pca.predict(X_pca_test)
```

```python
# Multinominal Lojistik Regresyon modeli iÃ§in confusion matrix'i hesaplama
conf_matrix_logistic_pca = confusion_matrix(y_test, logistic_reg_pca_preds)

# Sensitivity ve specificity deÄŸerlerini hesaplama
true_negatives_logistic_pca = conf_matrix_logistic_pca[0, 0]
false_negatives_logistic_pca = conf_matrix_logistic_pca[1, 0]
true_positives_logistic_pca = conf_matrix_logistic_pca[1, 1]
false_positives_logistic_pca = conf_matrix_logistic_pca[0, 1]

sensitivity_logistic_pca = true_positives_logistic_pca / (true_positives_logistic_pca + false_negatives_logistic_pca)
specificity_logistic_pca = true_negatives_logistic_pca / (true_negatives_logistic_pca + false_positives_logistic_pca)

# Multinominal Lojistik Regresyon modeli iÃ§in accuracy, recall ve f1-score hesaplama
accuracy_logistic_pca = accuracy_score(y_test, logistic_reg_pca_preds)
recall_logistic_pca = recall_score(y_test, logistic_reg_pca_preds)
f1_score_logistic_pca = f1_score(y_test, logistic_reg_pca_preds)

print("Multinominal Lojistik Regresyon (PCA):")
print("Sensitivity:", sensitivity_logistic_pca)
print("Specificity:", specificity_logistic_pca)
print("Accuracy:", accuracy_logistic_pca)
print("Recall:", recall_logistic_pca)
print("F1 Score:", f1_score_logistic_pca)
```

```python
# Veri setini X (Ã¶zellikler) ve y (etiketler) olarak tanÄ±mla
X = df.drop('X', axis=1)  # X deÄŸiÅŸkenini dÃ¼ÅŸÃ¼rerek Ã¶zellikleri belirtiyoruz
y = df['X']  # X deÄŸiÅŸkeni etiket (hedef deÄŸiÅŸken)

# Veriyi %70 eÄŸitim, %30 test olarak ayÄ±r
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

# Karar AÄŸacÄ± sÄ±nÄ±flandÄ±rma modelini oluÅŸtur
decision_tree = DecisionTreeClassifier(random_state=42)

# Modeli eÄŸit
decision_tree.fit(X_train, y_train)

# AÄŸaÃ§ yapÄ±sÄ±nÄ± gÃ¶rselleÅŸtir
plt.figure(figsize=(15,10))
tree.plot_tree(decision_tree, filled=True, feature_names=X.columns, class_names=['0', '1'])
plt.show()

# Test verisi Ã¼zerinde modeli deÄŸerlendirme
y_pred = decision_tree.predict(X_test)

# Performans metriklerini hesapla
accuracy = accuracy_score(y_test, y_pred)
precision = precision_score(y_test, y_pred)
recall = recall_score(y_test, y_pred)
f1 = f1_score(y_test, y_pred)
conf_matrix = confusion_matrix(y_test, y_pred)
class_report = classification_report(y_test, y_pred)

print("Accuracy:", accuracy)
print("Precision:", precision)
print("Recall:", recall)
print("F1 Score:", f1)
print("Confusion Matrix:\n", conf_matrix)
print("Classification Report:\n", class_report)
```

```python
# PCA modelini oluÅŸtur ve uygula
pca = PCA(n_components=2)  
X_pca = pca.fit_transform(X)

X_pca_train, X_pca_test, y_train, y_test = train_test_split(X_pca, y, test_size=0.3, random_state=42)

# Karar AÄŸacÄ± sÄ±nÄ±flandÄ±rma modelini oluÅŸtur
decision_tree_pca = DecisionTreeClassifier(random_state=42)

# Modeli eÄŸit
decision_tree_pca.fit(X_pca_train, y_train)

# AÄŸaÃ§ yapÄ±sÄ±nÄ± gÃ¶rselleÅŸtir
plt.figure(figsize=(15,10))
tree.plot_tree(decision_tree_pca, filled=True, class_names=['0', '1'])
plt.show()

# Test verisi Ã¼zerinde modeli deÄŸerlendirme
y_pred_pca = decision_tree_pca.predict(X_pca_test)

# Performans metriklerini hesapla
accuracy_pca = accuracy_score(y_test, y_pred_pca)
precision_pca = precision_score(y_test, y_pred_pca)
recall_pca = recall_score(y_test, y_pred_pca)
f1_pca = f1_score(y_test, y_pred_pca)
conf_matrix_pca = confusion_matrix(y_test, y_pred_pca)
class_report_pca = classification_report(y_test, y_pred_pca)

print("PCA Model Performance Metrics:")
print("Accuracy:", accuracy_pca)
print("Precision:", precision_pca)
print("Recall:", recall_pca)
print("F1 Score:", f1_pca)
print("Confusion Matrix:\n", conf_matrix_pca)
print("Classification Report:\n", class_report_pca)
```
